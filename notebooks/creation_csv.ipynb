```python
import os
import pandas as pd

json_file = "goodreads_books.json"

fra_file = "FRA.csv"
eng_file = "ENG.csv"
dut_file = "DUT.csv"

chunksize = 10000

for file in [fra_file, eng_file, dut_file]:
    if os.path.exists(file):
        os.remove(file)

for i, chunk in enumerate(pd.read_json(json_file, lines=True, chunksize=chunksize)):
    print(f"Processing chunk {i + 1}...")

    chunk["language_code_str"] = chunk["language_code"].fillna("").astype(str)

    fra_chunk = chunk[chunk["language_code_str"] == "fre"].drop(columns=["language_code_str"])
    eng_chunk = chunk[chunk["language_code_str"].str.startswith("en")].drop(columns=["language_code_str"])
    dut_chunk = chunk[chunk["language_code_str"] == "nl"].drop(columns=["language_code_str"])

    if not fra_chunk.empty:
        fra_chunk.to_csv(
            fra_file,
            mode="a",
            index=False,
            header=not os.path.exists(fra_file)
        )

    if not eng_chunk.empty:
        eng_chunk.to_csv(
            eng_file,
            mode="a",
            index=False,
            header=not os.path.exists(eng_file)
        )

    if not dut_chunk.empty:
        dut_chunk.to_csv(
            dut_file,
            mode="a",
            index=False,
            header=not os.path.exists(dut_file)
        )

print("Done.")
```

```python
import string
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import cross_validate
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import recall_score
import unicodedata
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.model_selection import train_test_split, cross_validate
```

```python
def preprocessing(sentence):

    lemmatizer = WordNetLemmatizer()

    # Basic cleaning
    sentence = sentence.strip()
    sentence = sentence.lower()
    sentence = ''.join(char for char in sentence if not char.isdigit())

    # Remove punctuation
    for punctuation in string.punctuation:
        sentence = sentence.replace(punctuation, '')

    # Tokenize
    tokenized_sentence = word_tokenize(sentence)

    # 1️⃣ Lemmatize verbs
    verb_lemmatized = [
        lemmatizer.lemmatize(word, pos="v")
        for word in tokenized_sentence
    ]

    # 2️⃣ Lemmatize nouns
    noun_lemmatized = [
        lemmatizer.lemmatize(word, pos="n")
        for word in verb_lemmatized
    ]

    # Join back
    cleaned_sentence = ' '.join(noun_lemmatized)

    return cleaned_sentence
```

```python
data = reviews_small[['rating','review_text']].copy()
data['clean_reviews'] = data["review_text"].apply(preprocessing)
data
```

```python
data['target'] = (data['rating'] > 3).astype(int)
data['review_text'][9995]
```

```python
count_vectorizer = CountVectorizer()
X = count_vectorizer.fit_transform(data['clean_reviews'])
X.toarray() # pour optimizé l'espace
```

```python
display(count_vectorizer.get_feature_names_out())
X.shape
```

```python
# Feature/Target
X = X
y = data['target']

# Pipeline vectorizer + Naive Bayes
model = MultinomialNB()

# Cross-validation
cv_results = cross_validate(model, X, y, cv = 5, scoring = ["accuracy"])
average_accuracy = cv_results["test_accuracy"].mean()
np.round(average_accuracy,2)
```

```python
np.unique(y,return_counts=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
model.fit(X_train,y_train)
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
def preprocessing(sentence):

    lemmatizer = WordNetLemmatizer()

    # Basic cleaning
    sentence = sentence.strip()
    sentence = sentence.lower()

    # Remove accents 
    sentence = ''.join(
        c for c in unicodedata.normalize('NFD', sentence)
        if unicodedata.category(c) != 'Mn'
    )

    # Remove numbers
    sentence = ''.join(char for char in sentence if not char.isdigit())

    # Remove punctuation
    sentence = sentence.translate(str.maketrans('', '', string.punctuation))

    # Tokenize
    tokenized_sentence = word_tokenize(sentence)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokenized_sentence_cleaned = [
        w for w in tokenized_sentence if w not in stop_words
    ]

    # Lemmatize verbs
    verb_lemmatized = [
        lemmatizer.lemmatize(word, pos="v")
        for word in tokenized_sentence_cleaned
    ]

    # Lemmatize nouns
    noun_lemmatized = [
        lemmatizer.lemmatize(word, pos="n")
        for word in verb_lemmatized
    ]

    # Join back
    cleaned_sentence = ' '.join(noun_lemmatized)

    return cleaned_sentence
```

```python
# data = reviews_small[['rating','n_votes','n_comments','riview_text']].copy()
data = reviews_small[['rating','review_text']].copy()
data.columns = ['score','message']
# data['score'].value_counts()
data = data[data.score<3]
data['clean_text'] = data["message"].apply(preprocessing)
data.sample(5)
```

```python
# Instantiating the TfidfVectorizer
tf_idf_vectorizer = TfidfVectorizer(
    ngram_range=(3,3),
    min_df=5,          # très important pour réduire la taille
    max_features=10000 # optionnel mais recommandé
)

# Training it on the texts
X = tf_idf_vectorizer.fit_transform(data['clean_text'])

# Créer un DataFrame SPARSE (pas dense !)
weighted_words = pd.DataFrame.sparse.from_spmatrix(
    X,
    columns=tf_idf_vectorizer.get_feature_names_out()
)
weighted_words
```

```python
# Instantiate the LDA
n_components = 3
lda_model = LatentDirichletAllocation(
    n_components=n_components, 
    max_iter = 100)

# Fit the LDA on the vectorized documents
lda_model.fit(weighted_words)
```

```python
document_topic_mixture = lda_model.transform(weighted_words)
document_topic_mixture
topic_word_mixture = pd.DataFrame(
    lda_model.components_,
    columns = tf_idf_vectorizer.get_feature_names_out()
)
topic_word_mixture
```

```python
def print_topics(lda_model, vectorizer, top_words):
    # 1. TOPIC MIXTURE OF WORDS FOR EACH TOPIC
    topic_mixture = pd.DataFrame(
        lda_model.components_,
        columns = vectorizer.get_feature_names_out()
    )

    # 2. FINDING THE TOP WORDS FOR EACH TOPIC
    ## Number of topics
    n_components = topic_mixture.shape[0]

    ## Top words for each topic
    for topic in range(n_components):
        print("-"*10)
        print(f"For topic {topic}, here are the the top {top_words} words with weights:")

        topic_df = topic_mixture.iloc[topic]\
            .sort_values(ascending = False).head(top_words)

        print(round(topic_df,3))

print_topics(lda_model, tf_idf_vectorizer, 10)
```
