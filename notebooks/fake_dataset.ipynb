```python
for _ in range(1000):
    en_fake = Faker(locale="en_US")
    country_name = en_fake.country()
    try:
        country_alpha_2 = pycountry.countries.get(name=country_name, "United States").alpha_2
        country = CountryInfo(country_alpha_2)
        print(country_name)
    except KeyError:
        print(f"Country '{country_name}' not found in pycountry database.")
```
